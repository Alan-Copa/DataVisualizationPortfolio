# Gravitational Constant

```python
import matplotlib.pyplot as plt
import numpy as np

# Data points for G measurements
# Year, G value, relative standard uncertainty (ppm), and category (based on description)
data = [
    (1969, 6.6732, 460, "recommended"),
    (1973, 6.6720, 730, "recommended"),
    (1986, 6.67449, 120, "recommended"),
    (1998, 6.673, 1500, "recommended"),
    (2002, 6.6742, 150, "recommended"),
    (2006, 6.67428, 100, "recommended"),
    (2010, 6.67384, 120, "recommended"),
    (2014, 6.67408, 46, "recommended"),
    (2018, 6.67430, 22, "recommended"),
    (2022, 6.67430, 22, "recommended"),
]

# Reference value for ppm calculation (CODATA 2014)
ppm_reference = 6.67408

# Calculate absolute uncertainty and ppm deviation
calculated_data = [
    (year, g_value, g_value * (uncertainty / 1e6), (g_value - ppm_reference) * 1e5, category)
    for year, g_value, uncertainty, category in data
]

# Separate data by category
categories = {"recommended": "green"}

# Prepare the plot
fig, ax1 = plt.subplots(figsize=(12, 8))

# Plot G values with error bars
for year, g_value, uncertainty, ppm_deviation, category in calculated_data:
    color = categories[category]
    ax1.errorbar(year, g_value, yerr=uncertainty, fmt='o', color=color, label=category if year == 1969 else "")

# Configure left y-axis for G values
ax1.set_xlabel("Year", fontsize=14)
ax1.set_ylabel("$G$ [$10^{-11}$ m$^3$ kg$^{-1}$ s$^{-2}$]", fontsize=14)
ax1.set_xlim(1960, 2030)
ax1.set_ylim(6.662, 6.680)  # Custom range for G values
ax1.set_yticks(np.arange(6.662, 6.680, 0.002))
ax1.grid(True, linestyle='--', alpha=0.7)

# Add a second y-axis for ppm deviation
ax2 = ax1.twinx()
ax2.set_ylabel("ppm (CODATA 2014)", fontsize=14)
ax2.set_ylim(-1800, 900)  # Custom range for ppm deviations
ax2.set_yticks(np.arange(-1800, 901, 300))

# Plot ppm deviations on the secondary y-axis
for year, g_value, uncertainty, ppm_deviation, category in calculated_data:
    ax2.plot(year, ppm_deviation, marker='x', color='red', label="ppm deviation" if year == 1969 else "")

# Add legend
handles = [
    plt.Line2D([0], [0], color="green", marker='o', linestyle='', label="Recommended values"),
    plt.Line2D([0], [0], color="red", marker='x', linestyle='', label="ppm deviation")
]
ax1.legend(handles=handles, title="Legend", loc="upper left")

# Title and layout
plt.title("Timeline of Measurements and Recommended Values for $G$ with ppm Deviations", fontsize=16)
plt.tight_layout()

# Show plot
plt.show()

# Save plot as pdf file
fig.savefig("gravitational_constant.pdf")


# Save plot as pdf file
fig.savefig("gravitational_constant.pdf")
```
