```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the datasets
ozone_consumption = pd.read_csv('ozone-depleting-substance-consumption.csv')
ozone_hole = pd.read_csv('antarctic-ozone-hole-area.csv')
stratospheric_ozone = pd.read_csv('stratospheric-ozone-concentration.csv')

# Clean the datasets
ozone_consumption_cleaned = ozone_consumption.fillna(0)
ozone_hole_cleaned = ozone_hole.interpolate()
stratospheric_ozone_cleaned = stratospheric_ozone.interpolate()

ozone_consumption_cleaned['Year'] = ozone_consumption_cleaned['Year'].astype(int)
ozone_hole_cleaned['Year'] = ozone_hole_cleaned['Year'].astype(int)
stratospheric_ozone_cleaned['Year'] = stratospheric_ozone_cleaned['Year'].astype(int)

# Prepare data
ozone_consumption_cleaned['Total Consumption'] = ozone_consumption_cleaned.iloc[:, 3:].sum(axis=1)
global_consumption = ozone_consumption_cleaned.groupby('Year')['Total Consumption'].sum()

# Scale data moderately
scaled_consumption = global_consumption / 1e6  # Scale down by one million for better visibility
scaled_hole_area = ozone_hole_cleaned['Maximum ozone hole area'] / 1e6  # Scale down by 1,000,000
scaled_ozone_concentration = stratospheric_ozone_cleaned['Minimum daily concentration']  # Leave as is

# Plot with multiple Y-axes
fig, ax1 = plt.subplots(figsize=(16, 10))

# Plot 1: Ozone consumption
ax1.plot(global_consumption.index, scaled_consumption.values, color='red', label='Ozone-Depleting Substance Consumption')
ax1.set_xlabel('Year', fontsize=14)
ax1.set_ylabel('Ozone-Depleting Substance Consumption (Million Tons)', color='red', fontsize=14)  # Updated label with units
ax1.tick_params(axis='y', labelcolor='red', labelsize=12)
ax1.tick_params(axis='x', labelsize=12)
ax1.grid(True)

# Add second Y-axis for Ozone Hole Area
ax2 = ax1.twinx()
ax2.plot(ozone_hole_cleaned['Year'], scaled_hole_area, color='blue', label='Maximum Ozone Hole Area')
ax2.set_ylabel('Ozone Hole Area (Million km²)', color='blue', fontsize=14)
ax2.tick_params(axis='y', labelcolor='blue', labelsize=12)

# Add third Y-axis for Ozone Concentration
ax3 = ax1.twinx()
ax3.spines['right'].set_position(('outward', 80))  # Offset third axis
ax3.plot(stratospheric_ozone_cleaned['Year'], scaled_ozone_concentration, color='green', label='Minimum Ozone Concentration')
ax3.set_ylabel('Ozone Concentration (Dobson Units)', color='green', fontsize=14)
ax3.tick_params(axis='y', labelcolor='green', labelsize=12)

# Add legend inside the grid at the top-right corner
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
lines_3, labels_3 = ax3.get_legend_handles_labels()
all_lines = lines_1 + lines_2 + lines_3
all_labels = labels_1 + labels_2 + labels_3
ax1.legend(all_lines, all_labels, loc='upper right', fontsize=12, frameon=True, edgecolor='black', facecolor='white')

# Title and show
plt.title('Trends: Ozone Depletion Substances, Hole Area, and Concentration (1979-2021)', fontsize=18)
plt.tight_layout()
plt.show()

# save the plot as a pdf file
fig.savefig('ozone_data_visualization_t4.pdf')
```
